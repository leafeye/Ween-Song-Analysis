```python
#selecting albums and IDS manually
albums = ['God Ween Satan: The Oneness', 'The Pod', 'Pure Guava', 'Chocolate and Cheese',
 '12 Golden Country Greats', 'The Mollusk', 'White Pepper', 'Quebec',
 'Shinola Vol. 1', 'La Cucaracha']

album_ids = ['3vvn6nTmlQd2JBABF8Hikb', '7tNajQI0M0ZdKTmzkj84Z7', '6jr7huVsu2VbhvBvyRSvLN', '6IsHMRIL7pVXmrg4b2ssb0',
 '6H5fUQk4FkT9iR4SNHnK3D', '1yfJqxKKXG320vhqLfUEeC', '6xHEzpqk0yDUt2uuodhWqF', '29uDLfGfPssBvHcHaWDddl',
 '0ijuPyfyaNJbVNwGQm6Xex', '3JUHw8diOk5Y0Z6w44rVyS']
```

```python

for i, album in enumerate(albums):
    alb_name = album
    alb_id = album_ids[i]
    
    #lists for dataframe
    nr_l = []
    title_l = []
    duration_l = []
    acousticness_l = []
    danceability_l = []
    energy_l = []
    instrumentalness_l = []
    liveness_l = []
    loudness_l = []
    speechiness_l = []
    mode_l = []
    tempo_l = []
    time_signature_l = []
    valence_l = []

    #tracks
    alb_tracks = sp.album_tracks(alb_id)

    for j, track in enumerate(alb_tracks['items']):
        track_nr = j+1
        track_title = track['name']
        track_id = track['id']

        nr_l.append(track_nr)
        title_l.append(track_title)


        #features
        track_feat = sp.audio_features(track_id)[0]

        #append features to the lists
        duration_l.append(track_feat['duration_ms'])
        acousticness_l.append(track_feat['acousticness'])
        danceability_l.append(track_feat['danceability'])
        energy_l.append(track_feat['energy'])
        instrumentalness_l.append(track_feat['instrumentalness'])
        liveness_l.append(track_feat['liveness'])
        loudness_l.append(track_feat['loudness'])
        speechiness_l.append(track_feat['speechiness'])
        mode_l.append(track_feat['mode'])
        tempo_l.append(track_feat['tempo'])
        time_signature_l.append(track_feat['time_signature'])
        valence_l.append(track_feat['valence'])

    #for first album (i == 0), create the feature dataframe
    if i == 0:
        feat_df = pd.DataFrame({'album': alb_name,
                           'track_nr': nr_l,
                           'track_title': title_l,
                           'duration_ms': duration_l,
                           'acousticness': acousticness_l,
                           'danceability': danceability_l,
                           'energy': energy_l,
                           'instrumentalness': instrumentalness_l,
                           'liveness': liveness_l,
                           'loudness': loudness_l,
                           'speechiness': speechiness_l,
                           'mode': mode_l,
                           'tempo': tempo_l,
                           'time_signature': time_signature_l,
                           'valence': valence_l,
                      })
    #for next albums, create per album a dataframe and append to the feature dataframe
    else:
        album_df = pd.DataFrame({'album': alb_name,
                           'track_nr': nr_l,
                           'track_title': title_l,
                           'duration_ms': duration_l,
                           'acousticness': acousticness_l,
                           'danceability': danceability_l,
                           'energy': energy_l,
                           'instrumentalness': instrumentalness_l,
                           'liveness': liveness_l,
                           'loudness': loudness_l,
                           'speechiness': speechiness_l,
                           'mode': mode_l,
                           'tempo': tempo_l,
                           'time_signature': time_signature_l,
                           'valence': valence_l,
                      })

        feat_df = feat_df.append(album_df, ignore_index=True)   
```
